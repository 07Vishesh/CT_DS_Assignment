# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()

    try:
      if "calculate" in query_lower:
        expression=query_lower.replace("calculate","").strip()
        return{type:"calculation","result":calculator(expression)}
      elif "keyword" in query_lower:
        text=(query.replace("Extract keywords from","").replace("extract keywords from","").replace("keywords","").strip())
        return{"type":"keywords","result":extract_keywords(text)}
      else:
        return{"type":"general","result":f"I understand your question:'{query}'. This is a general query and does not require any tool."}
    except Exception as e:
      return{"type":"error","result":str(e)}
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
import logging
import datetime

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)
def word_count(text):
    return len(text.split())
def text_reverse(text):
    return text[::-1]
def current_time():
    return datetime.datetime.now().strftime("%d-%m-%Y %H:%M:%S")

def agent(query: str):
    query_lower = query.lower()
    logging.info(f"Received Query: {query}")
    try:
        if any(word in query_lower for word in ["calculate", "add", "subtract", "multiply", "divide"]):
            expression = query_lower.replace("calculate", "").strip()
            logging.info("Routing -> Calculator Tool")
            return {
                "tool": "Calculator",
                "status": "success",
                "result": calculator(expression)
            }

        elif "keyword" in query_lower:
            text = (
                query
                .replace("Extract keywords from", "")
                .replace("extract keywords from", "")
                .replace("keywords", "")
                .strip()
            )
            logging.info("Routing -> Keyword Extractor")
            return {
                "tool": "Keyword Extractor",
                "status": "success",
                "result": extract_keywords(text)
            }

        elif "count words" in query_lower:
            text = query_lower.replace("count words", "").strip()
            logging.info("Routing -> Word Count")
            return {
                "tool": "Word Counter",
                "status": "success",
                "result": word_count(text)
            }

        elif "reverse" in query_lower:
            text = query.replace("reverse", "").strip()
            logging.info("Routing -> Reverse Tool")
            return {
                "tool": "Reverse Text",
                "status": "success",
                "result": text_reverse(text)
            }

        elif "time" in query_lower:
            logging.info("Routing -> Time Tool")
            return {
                "tool": "Time",
                "status": "success",
                "result": current_time()
            }

        else:
            logging.info("Routing -> General Response")
            return {
                "tool": "General",
                "status": "success",
                "result": f"I understood your query: '{query}'. No specialized tool was required."
            }

    except Exception as e:
        logging.error(str(e))
        return {
            "tool": "Error",
            "status": "failed",
            "result": str(e)
        }
````

````python
queries = [
    "Calculate 30*15",
    "Extract keywords from Artificial Intelligence is transforming healthcare and education",
    "Count words Machine learning is amazing",
    "Reverse Palindrome",
    "What is the consciousness?",
    "Who invented JavaScript?"
]
for q in queries:
    print("="*60)
    print("Query :", q)
    print(agent(q))
````

````python
while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
